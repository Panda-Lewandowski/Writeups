Дается прямоугольный параллелепипед 6x6x36 из символов, состоящий из плоскостей, слепленных вместе. В каждой плоскости символы не должны повторяться. На линиях, идущих вглубь параллелепипеда сквозь элементы плоскости символы не должны повторяться. Некоторых символов не хватает (они обозначаются *), существует только одно возможное решение. Флаг состоит из символов на первой плоскости.

```python
import string
import itertools
import copy
import pprint
```

```python
alphabet = []
unknown = {}
with open('ppc.txt') as file:
    blocks = []
    chunk = []
    for line in file:
        if line.find("Field") == -1 and line != '\n':
            sym_list = line[:-2].split(" ")
            print(sym_list)
            for i_sym, sym in enumerate(sym_list):
                if sym == "*":
                    if not unknown.get(len(blocks), None):
                        unknown[len(blocks)] = {len(chunk): {i_sym: None}}
                    else:
                        if not unknown.get(len(blocks), {}).get(len(chunk), None):
                            unknown[len(blocks)][len(chunk)] = {i_sym: None}
                        else:
                            unknown[len(blocks)][len(chunk)][i_sym] = None
                else:
                    if sym not in alphabet:
                        alphabet.append(sym)
            chunk.append(sym_list)
                
        else: 
            if chunk:
                blocks.append(chunk)
            chunk = []
            print("----")
            
print(alphabet, len(alphabet))
unknown
```

```python
%%time
for kBl in unknown:
    for kCh in unknown[kBl]:
        for kSym in unknown[kBl][kCh]:
            unknown[kBl][kCh][kSym] = copy.copy(alphabet)

while len(unknown) != 0:
    foundBl = []
    for kBl in unknown:
    #     print(kBl)
        for ch in blocks[kBl]:
            for sym in ch:
    #             print(sym, end="")
                foundCh = []
                for kCh in unknown[kBl]:
                    foundSym = []
                    for kSym in unknown[kBl][kCh]:
                        
                        if sym in unknown[kBl][kCh][kSym]:
                            unknown[kBl][kCh][kSym].remove(sym)
                        for bl in blocks[:kBl] + blocks[kBl+1:]:
                            VGLUB_BLYA = bl[kCh][kSym]
                            if VGLUB_BLYA in unknown[kBl][kCh][kSym]:
                                unknown[kBl][kCh][kSym].remove(VGLUB_BLYA)
                        if len(unknown[kBl][kCh][kSym]) == 1:
                            blocks[kBl][kCh][kSym] = unknown[kBl][kCh][kSym][0]
                            foundSym.append(kSym)
                    for kSym in foundSym:
                        unknown[kBl][kCh].pop(kSym)
                    if len(unknown[kBl][kCh]) == 0:
                        foundCh.append(kCh)
                for kCh in foundCh:
                    unknown[kBl].pop(kCh)
        if len(unknown[kBl]) == 0:
            foundBl.append(kBl)
    for kBl in foundBl:
        unknown.pop(kBl)
            
            
                            
#     print("\n")
#     print(unknown[kBl])

blocks
```
